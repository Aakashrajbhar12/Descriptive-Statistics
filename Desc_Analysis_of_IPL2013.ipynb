#### Visualizations --Seaborn

```python
import seaborn as sns
```

```python
#Stripplot--1st
sns.stripplot(y='WKTS', data=IPL_2013, jitter=True)
plt.show()
```

```python
#Stripplot --2nd 
sns.stripplot(x="AGE", y='WKTS', data=IPL_2013, jitter=False)
plt.show()
```

```python
#Swarmplot --1st
sns.swarmplot(x="AGE", y='WKTS', data=IPL_2013)
plt.show()
```

```python
#Boxplot
sns.boxplot(x="AGE", y='WKTS', data=IPL_2013)
plt.show()
```

```python
#Violinplot
sns.violinplot(x="AGE", y='WKTS', data=IPL_2013)
plt.show()
```

```python
#Combine plot
plt.subplot(1,2,1)
sns.boxplot(x="AGE", y='WKTS', data=IPL_2013)

plt.subplot(1,2,2)
sns.violinplot(x="AGE", y='WKTS', data=IPL_2013)
plt.show()
```

```python
#Combine plot --2nd
sns.violinplot(x="AGE", y='WKTS', data=IPL_2013, inner=None, color="grey")
sns.stripplot(x="AGE", y='WKTS', data=IPL_2013, jitter=True)
plt.show()
```

```python
#JointPlot
sns.jointplot(x= 'AGE', y= 'WKTS', data=IPL_2013)
plt.show()
```

```python
#Density plot
sns.jointplot(x= 'AGE', y= 'WKTS', data=IPL_2013, kind="kde")
plt.show()
```
